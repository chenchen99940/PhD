```python
import numpy as np
import matplotlib.pyplot as plt
p=0.8;
r=0.2;
def a(l):
    return (r-r*(p-r)**l)/(1+r-p);
def b(l):
    return (r+(1-p)*(p-r)**l)/(1+r-p);

def whittle(l):
    return ((a(l)-a(l+1))*(l+1)+a(l+1))/(1-p+(a(l)-a(l+1))*l+a(l+1));

x = np.arange(0, 25, 1);
y = a(x);
z = b(x);
w = whittle(x);

plt.subplot(1,2,1);
plt.plot(x, y);

plt.subplot(1,2,1);
plt.plot(x,z);

plt.subplot(1,2,2);
plt.plot(x,w);

plt.subplot(1,2,2);
plt.axhline(y= r/((1-p)*(1+r-p)+r), color='r', linestyle='-')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.path as mpath

p=0.9;
r=1/15;

q=0.5;
s=1/3;

def a(l):
    return (r-r*(p-r)**l)/(1+r-p);
def b(l):
    return (r+(1-p)*(p-r)**l)/(1+r-p);
def whittleone(l):
    return ((a(l)-a(l+1))*(l+1)+a(l+1))/(1-p+(a(l)-a(l+1))*l+a(l+1));

def c(l):
    return (s-s*(q-s)**l)/(1+s-q);
def d(l):
    return (s+(1-q)*(q-s)**l)/(1+s-q);
def whittletwo(l):
    return ((c(l)-c(l+1))*(l+1)+c(l+1))/(1-q+(c(l)-c(l+1))*l+c(l+1));

x = np.arange(0, 20, 1);
y = a(x);
z = b(x);
w=whittleone(x);
u= c(x);
v= d(x);
r= whittletwo(x);


star = mpath.Path.unit_regular_star(6)
circle = mpath.Path.unit_circle()
# concatenate the circle with an internal cutout of the star
verts = np.concatenate([circle.vertices, star.vertices[::-1, ...]])
codes = np.concatenate([circle.codes, star.codes])
cut_star = mpath.Path(verts, codes)

fig = plt.figure()
fig, ax = plt.subplots();
ax.plot(x, y, '--r', label='Class One',marker=cut_star, markersize=5)
ax.plot(x, z, '--r',marker=cut_star, markersize=5)
ax.plot(x, u, ':g', label='Class Two',marker=cut_star, markersize=5)
ax.plot(x, v, ':g',marker=cut_star, markersize=5)
plt.ylabel('Belief value')
plt.xlabel('Time elapsed since last observation')



legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#8c564b')


plt.savefig('fig1.jpg',bbox_inches='tight')




```
