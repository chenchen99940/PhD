```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N2=np.array([100,120,140,200,400,600,800,1000,1500])
gap2=np.array([2.5,2.06,1.75,1.13,0.52,0.32,0.24,0.2,0.06])
```

```python
plt.plot(N2,gap2)
```

```python
N1=np.array([100,200,1000,1500,2000,5000])
gap1=np.array([4,3.96,3.31,2.91,2.61,1.92])
plt.plot(N1,gap1)
```

```python
plt.plot(N1,gap1)
plt.plot(N2,gap2)
plt.legend(['sub-optimality gap for model 1',
           'sub-optimality gap for model 2',
           'Curve 200/N'])
plt.xlim([0,1500])
#plt.xlim([0,5000])
```

```python
plt.plot(N2,gap2)
plt.plot(N2,200/N2,'--')
plt.legend(['sub-optimality gap for model 2',
           r'$200 / N$'])
plt.xlabel('N')
plt.ylabel('suboptimality gap')
```

```python
E = [1,1,1,1,2,1,.5,1,1]
plt.errorbar(N2,gap2,E)
```
